–í—Ö–æ–¥–Ω—ã–µ –¥–∞–Ω–Ω—ã–µ
–í –ø–µ—Ä–≤–æ–π —Å—Ç—Ä–æ–∫–µ –∑–∞–ø–∏—Å–∞–Ω–æ —Ü–µ–ª–æ–µ —á–∏—Å–ª–æ ùëõ (3‚â§ùëõ‚â§2‚ãÖ105) ‚Äî –∫–æ–ª–∏—á–µ—Å—Ç–≤–æ –≥–æ—Ä–æ–¥–æ–≤ –≤ –ó–æ–ª–æ—Ç–æ–º –∫–æ–ª—å—Ü–µ –ë–µ—Ä–ª—è–Ω–¥–∏–∏.

–ì–∞—Ä–∞–Ω—Ç–∏—Ä—É–µ—Ç—Å—è, —á—Ç–æ —Ö–æ—Ç—è –±—ã –æ–¥–Ω–æ –∏–∑ —á–∏—Å–µ–ª ùëéùëñ —Å—Ç—Ä–æ–≥–æ –±–æ–ª—å—à–µ –Ω—É–ª—è.

–í—ã—Ö–æ–¥–Ω—ã–µ –¥–∞–Ω–Ω—ã–µ

```python
def task1(n, a):
    a.reverse()
    max_a=max(a)
    otv=n*max_a - a.index(max(a))
    return(otv)

print('–í—Ö–æ–¥–Ω—ã–µ –¥–∞–Ω–Ω—ã–µ:')
n = int(input())
a = list(map(int, input().split()))
if (n<=3 and n>=100000):
    raise ValueError('–û—à–∏–±–∫–∞ –≤–≤–µ–¥–µ–Ω–∏—è —á–∏—Å–ª–∞ n')
else:
    for i in range(n):
        if (a[i]<=0 or a[i]>=1000000000):
            raise ValueError('–û—à–∏–±–∫–∞ –≤–≤–µ–¥–µ–Ω–∏—è —Å—Ç—Ä–æ–∫–∏ —Å —á–∏—Å–ª–∞–º–∏ ùëéùëñ')
            break
    print('–í—ã—Ö–æ–¥–Ω—ã–µ –¥–∞–Ω–Ω—ã–µ: ', task1(n,a))
```

–í—ã—Ö–æ–¥–Ω—ã–µ –¥–∞–Ω–Ω—ã–µ
–í –ø–µ—Ä–≤—É—é —Å—Ç—Ä–æ–∫—É –≤—ã–≤–µ–¥–∏—Ç–µ —Ü–µ–ª–æ–µ —á–∏—Å–ª–æ z ‚Äî –∫–æ–ª–∏—á–µ—Å—Ç–≤–æ –∏—Å–∫–æ–º—ã—Ö –ø–∞—Ä. –î–∞–ª–µ–µ –≤—ã–≤–µ–¥–∏—Ç–µ –≤—Å–µ –∏—Å–∫–æ–º—ã–µ –ø–∞—Ä—ã (a,‚Äâb) –≤ –ø–æ—Ä—è–¥–∫–µ —É–≤–µ–ª–∏—á–µ–Ω–∏—è –ø–µ—Ä–≤–æ–π –∫–æ–º–ø–æ–Ω–µ–Ω—Ç—ã a.

```python
def task9(n,g):
    subtree_size=[0]*n
    stack=[[-1,0,0]]
    ans=[]
    while stack:
        par,ver,state=stack.pop()
        if state==0:
            stack.append([par,ver,1])
            for to in g[ver]:
                if to!=par:
                    stack.append([ver,to,0])
        else:
            if len(g[ver])==1:
                subtree_size[ver]=1
            else:
                cnt=0
                tmp=[]
                for to in g[ver]:
                    if to!=par:
                        cnt+=subtree_size[to]
                        tmp.append(subtree_size[to])
                tmp.append(n-cnt-1)
                local=[0]*(n+1)
                local[0]=1
                for x in tmp:
                    for i in range(n-x,-1,-1):
                        if local[i]==1:
                            local[i+x]=1
                            if x+i!=0 and n-1-(x+i)!=0:
                                ans.append(x+i)
                subtree_size[ver]=cnt+1
    return(sorted(list(set(ans))))
print('–í—Ö–æ–¥–Ω—ã–µ –¥–∞–Ω–Ω—ã–µ:')
n=int(input())
if(n>5000 or n<3):
    raise ValueError('–û—à–∏–±–∫–∞ –≤–≤–µ–¥–µ–Ω–∏—è —á–∏—Å–ª–∞ n')
else:
    g=[[] for i in range(n)]
    for i in range(n-1):
        x,y=map(int,input().split())
        if(x<1):
            raise ValueError('–û—à–∏–±–∫–∞ –≤–≤–µ–¥–µ–Ω–∏—è —á–∏—Å–ª–∞ x')
        if(y>n):
            raise ValueError('–û—à–∏–±–∫–∞ –≤–≤–µ–¥–µ–Ω–∏—è —á–∏—Å–ª–∞ y')
        x-=1
        y-=1
        g[x].append(y)
        g[y].append(x)
    ans=task9(n,g)
    print('–í—ã—Ö–æ–¥–Ω—ã–µ –¥–∞–Ω–Ω—ã–µ:')
    print(len(ans))
    for x in ans:
        print(x,n-1-x)
```
